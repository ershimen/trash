```python
def print_grid2(grid, h, w):
    print("\n\t+", "---+"*w, sep='')
    for i in range(h):
        print("\t|", end='')
        for j in range(w):
            if grid[i][j] == True:
                print(" # |", end='')
            elif grid[i][j] == False:
                print(" · |", end='')
            else:
                print("   |", end='')
        print("\n\t+", "---+"*w, sep='')
```

```python
def cabe2(arr, i, n):
    if  i + n > len(arr): # Si no hay espacio (se puede omitir?)
        return (False, -1)
    if i > 0 and arr[i-1]: # Si tiene un True detras
        return (False, -1)
    index = i
    left = n
    while left != 0 and index < len(arr):
        #print("\t\tarr[%d]: " % index, arr[index], ", left: %d" % left, sep='')
        if arr[index] == False:
            left = n
        else:
            left = left - 1
        index = index + 1
    if index < len(arr) and arr[index]: # Si el siguiente es true
        return cabe2(arr, i+1, n)
    #print("arr[%d]:" % index, arr[index])
    if left == 0:
        if index < len(arr):
            return (arr[index] != True, index - n)
        return (True, index - n)
    return (False, index - n)

def cabe2(arr, i, n):
    if i > 0 and arr[i-1]: # Si tiene un True detras
        return False
    index = i
    left = n
    while left != 0 and index < len(arr): # Ver si cabe
        if arr[index] == False:
            return False
        left = left - 1
        index = index + 1
    if left > 0 or (index < len(arr) and arr[index]): # Si tiene un True delante
        return False
    return True
```

```python
def combinations2(res, sol, sol_i, vals, vals_i, n):
    if len(vals) == 1 and vals[0] == 0: # si se buscan combinaciones de 0 conjuntos
        fill(sol)
        res.append(sol)
        return res
    if vals_i == len(vals): # Si se han insertado todas las secuencias, añadir al resultado
        fill(sol)
        # Comprobar si hay mas True
        index = sol_i
        while index < len(sol):
            if sol[index]:
                return res
            index = index + 1
        res.append(sol)
        return res
    copy = sol.copy()
    while True:
        bool_cabe, index_cabe = cabe2(copy, sol_i, vals[vals_i])
        #print("arr:", copy, ", i:", sol_i, ", n:", vals[vals_i], ", cabe:", bool_cabe, ", index:", index_cabe)
        if not bool_cabe:
            break
        sol_i = index_cabe
        for x in range(vals[vals_i]): # Añadir vals[vals_i] casillas en negro
            copy[sol_i+x] = True
        sol_i = sol_i + 1 # dejar un espacio en blanco
        # probar a insertar la siguiente secuencia o retornar
        combinations2(res, copy, sol_i + vals[vals_i], vals, vals_i+1, n)
        copy = sol.copy() # resetear el array cuando se prueba otra combinacion
    return res


def combinations2(res, sol, sol_i, vals, vals_i):
    if len(vals) == 1 and vals[0] == 0: # si se buscan combinaciones de 0 conjuntos
        fill(sol)
        res.append(sol)
        return res
    if vals_i == len(vals): # Si ya se han insertado todas las secuencias
        index = sol_i
        while index < len(sol): # Buscar a ver si hay mas True
            if sol[index] == True:
                return res
            index = index + 1
        fill(sol)
        res.append(sol)
        return res
    one_true = False
    while sol_i < len(sol):
        copy = sol.copy()
        cabe = cabe2(copy, sol_i, vals[vals_i])
        if copy[sol_i]: # Si hay un True significa que vals[vals_i] tiene que estar aqui
            one_true = True
        if cabe: # Si cabe, rellenar y buscar con vals[vals_i+1]
            for x in range(vals[vals_i]):
                copy[sol_i+x] = True
            sol_i = sol_i + 1 # Espacio en blanco
            combinations2(res, copy, sol_i + vals[vals_i], vals, vals_i + 1)
        else: # Si no cabe y habia un True, entonces la secuencia no es valida
            if one_true:
                return res
            sol_i = sol_i + 1
    return res
```

```python
def solve2(tablero, vals):
    rows = tablero.shape[0]
    cols = tablero.shape[1]
    turno = 0 # turno de fila
    unchanged = 0
    total_unchanged = rows + cols
    #print(total_unchanged)
    while unchanged < total_unchanged:
        #print("########################## Turno de: %d ##############################" % turno)
        for row_col in range(tablero.shape[turno]): # Por cada fila/columna
            if turno == 0:
                new_combinations = combinations2([], np.copy(tablero[row_col]), 0, vals[turno][row_col], 0, tablero.shape[1])
            else:
                new_combinations = combinations2([], np.copy(tablero[:, row_col]), 0, vals[turno][row_col], 0, tablero.shape[0])
            #print("new_combinations")
            #print_grid2(new_combinations, len(new_combinations), len(new_combinations[0]))
            #print("new_values")
            new_values = and_grid(new_combinations)
            #print_grid2([new_values], 1, len(new_values))
            if turno == 0:
                if np.sum(tablero[row_col] == new_values) == len(new_values):
                    unchanged = unchanged + 1
                else:
                    unchanged = 0
                vector_and(tablero[row_col], new_values)
                #print("vector_and")
                #print_grid2([tablero[row_col]], 1, len(tablero[row_col]))
            else:
                if np.sum(tablero[:, row_col] == new_values) == len(new_values):
                    unchanged = unchanged + 1
                else:
                    unchanged = 0
                vector_and(tablero[:, row_col], new_values)
                #print("vector_and")
                #print_grid2([tablero[:, row_col]], 1, len(tablero[:, row_col]))
            #print("tablero, unchanged:", unchanged)
            #print_grid2(tablero, rows, cols)
        turno = (turno + 1) % 2 # Cambio de turno
        #print_grid2(tablero, rows, cols)

def solve2(row_vals, col_vals):
    tablero = np.array([[None for _ in range(len(col_vals))]]*len(row_vals), dtype=object)
    turno = 0 # Turno de fila
    unchanged = 0 # Numero de iteraciones sin cambios en el tablero
    total_unchanged = tablero.shape[0] + tablero.shape[1] # rows + cols
    end = False
    while not end:
        for row_col in range(tablero.shape[turno]): # Por cada fila/columna
            # Obtener combinaciones
            if turno == 0:
                new_combinations = combinations2([], np.copy(tablero[row_col]), 0, row_vals[row_col], 0)
            else:
                new_combinations = combinations2([], np.copy(tablero[:, row_col]), 0, col_vals[row_col], 0)
            if len(new_combinations) == 0: # Si no tiene solucion
                return None
            # Calcular "and_grid"
            new_values = and_grid(new_combinations)
            # Guardar el resultado en el tablero
            if turno == 0:
                if np.sum(tablero[row_col] == new_values) == len(new_values):
                    unchanged = unchanged + 1
                else:
                    unchanged = 0
                vector_and(tablero[row_col], new_values)
            else:
                if np.sum(tablero[:, row_col] == new_values) == len(new_values):
                    unchanged = unchanged + 1
                else:
                    unchanged = 0
                vector_and(tablero[:, row_col], new_values)
            if unchanged >= total_unchanged: # Si no se pueden rellenar mas casillas
                end = True
                break
        turno = (turno + 1) % 2 # Cambio de turno
    return tablero
```

```python
def combinations3(res, sol, sol_i, vals, vals_i):
    if len(vals) == 1 and vals[0] == 0: # si se buscan combinaciones de 0 conjuntos
        fill(sol)
        #res.append(sol)
        grid_and(res, sol)
        return res
    if vals_i == len(vals): # Si ya se han insertado todas las secuencias
        index = sol_i
        while index < len(sol): # Buscar a ver si hay mas True
            if sol[index] == True:
                return res
            index = index + 1
        fill(sol)
        #res.append(sol)
        grid_and(res, sol)
        return res
    one_true = False
    while sol_i < len(sol):
        copy = sol.copy()
        cabe = cabe2(copy, sol_i, vals[vals_i])
        if copy[sol_i]: # Si hay un True significa que vals[vals_i] tiene que estar aqui
            one_true = True
        if cabe: # Si cabe, rellenar y buscar con vals[vals_i+1]
            for x in range(vals[vals_i]):
                copy[sol_i+x] = True
            sol_i = sol_i + 1 # Espacio en blanco
            combinations2(res, copy, sol_i + vals[vals_i], vals, vals_i + 1)
        else: # Si no cabe y habia un True, entonces la secuencia no es valida
            if one_true:
                return res
            sol_i = sol_i + 1
    return res
```

```python
def solve2(row_vals, col_vals):
    tablero = np.array([[None for _ in range(len(col_vals))]]*len(row_vals), dtype=object)
    turno = 0 # Turno de fila
    unchanged = 0 # Numero de iteraciones sin cambios en el tablero
    total_unchanged = tablero.shape[0] + tablero.shape[1] # rows + cols
    end = False
    while not end:
        for row_col in range(tablero.shape[turno]): # Por cada fila/columna
            # Obtener combinaciones
            if turno == 0:
                new_combinations = combinations2([], np.copy(tablero[row_col]), 0, row_vals[row_col], 0)
            else:
                new_combinations = combinations2([], np.copy(tablero[:, row_col]), 0, col_vals[row_col], 0)
            if len(new_combinations) == 0: # Si no tiene solucion
                return None
            # Calcular "and_grid"
            new_values = and_grid(new_combinations)
            # Guardar el resultado en el tablero
            if turno == 0:
                if np.sum(tablero[row_col] == new_values) == len(new_values):
                    unchanged = unchanged + 1
                else:
                    unchanged = 0
                vector_and(tablero[row_col], new_values)
            else:
                if np.sum(tablero[:, row_col] == new_values) == len(new_values):
                    unchanged = unchanged + 1
                else:
                    unchanged = 0
                vector_and(tablero[:, row_col], new_values)
            if unchanged >= total_unchanged: # Si no se pueden rellenar mas casillas
                end = True
                break
        turno = (turno + 1) % 2 # Cambio de turno
    return tablero
```
